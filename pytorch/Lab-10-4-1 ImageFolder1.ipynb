# Lab-10-4-1 ImageFolder (1)

```python
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
from matplotlib.pyplot import imshow
```

```python
trans = transforms.Compose([
    transforms.Resize((64,64))
])

train_data = torchvision.datasets.ImageFolder(
    root='images/cat_dog/train', transform=trans)

test_data = torchvision.datasets.ImageFolder(
    root='images/cat_dog/test', transform=trans)
```

```python
for num, value in enumerate(train_data):
    data, label = value
    imshow(data)
    break
```

```python
for num, value in enumerate(train_data):
    data, label = value
    if(label == 0):
        data.save('images/cat_dog/train_resized/cat/%d_%d.jpeg'%(num, label))
    else:
        data.save('images/cat_dog/train_resized/dog/%d_%d.jpeg'%(num, label))
```

```python
for num, value in enumerate(test_data):
    data, label = value
    if(label == 0):
        data.save('images/cat_dog/test_resized/cat/%d_%d.jpeg'%(num, label))
    else:
        data.save('images/cat_dog/test_resized/dog/%d_%d.jpeg'%(num, label))
```
